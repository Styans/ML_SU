```python
# === Lab & Practical: K-Means & DBSCAN ===
# Импорт библиотек
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, adjusted_rand_score
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

# Фиксируем случайность
np.random.seed(42)

# === Синтетические данные ===
# 1. Простые кластеры (blobs)
X_blobs, y_blobs = make_blobs(n_samples=800, centers=4, cluster_std=0.70, random_state=42)
Xb = StandardScaler().fit_transform(X_blobs)

# 2. Нелинейные данные (moons)
X_moons, y_moons = make_moons(n_samples=800, noise=0.06, random_state=42)
Xm = StandardScaler().fit_transform(X_moons)
```

```python
# === K-Means Clustering (Elbow method) ===
K_range = range(2, 11)
inertias = []

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(Xb)
    inertias.append(km.inertia_)

plt.figure()
plt.plot(list(K_range), inertias, marker='o')
plt.xlabel('K')
plt.ylabel('Inertia')
plt.title('Elbow Method — KMeans')
plt.show()
```

```python
# === Silhouette Score for different K ===
sil = []
for k in K_range:
    labels = KMeans(n_clusters=k, random_state=42).fit_predict(Xb)
    sil.append(silhouette_score(Xb, labels))

plt.figure()
plt.plot(list(K_range), sil, marker='o')
plt.xlabel('K')
plt.ylabel('Silhouette')
plt.title('Silhouette vs K')
plt.show()
```

```python
# === Финальная модель K-Means ===
best_K = 4
kmeans = KMeans(n_clusters=best_K, random_state=42)
labels_km = kmeans.fit_predict(Xb)

plt.figure()
plt.scatter(Xb[:,0], Xb[:,1], c=labels_km, s=10)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker='x', s=200, color='red')
plt.title(f'K-Means (K={best_K})')
plt.show()

# Оценка качества
ari_km = adjusted_rand_score(y_blobs, labels_km)
print("ARI (K-Means):", round(ari_km, 4))
```

```python
# === DBSCAN: k-distance график для подбора eps ===
min_samples = 5
nbrs = NearestNeighbors(n_neighbors=min_samples).fit(Xm)
distances, _ = nbrs.kneighbors(Xm)
k_dists = np.sort(distances[:, -1])

plt.figure()
plt.plot(k_dists)
plt.xlabel('Points sorted')
plt.ylabel('k-distance')
plt.title('k-distance plot (for eps)')
plt.show()
```

```python
# === DBSCAN clustering ===
eps = 0.3  # можно изменить, если кластеры слипаются
db = DBSCAN(eps=eps, min_samples=min_samples)
labels_db = db.fit_predict(Xm)

plt.figure()
plt.scatter(Xm[:,0], Xm[:,1], c=labels_db, s=10)
plt.title(f'DBSCAN (eps={eps})')
plt.show()

# Метрики
valid = labels_db != -1  # убираем шум
sil_db = silhouette_score(Xm[valid], labels_db[valid]) if np.unique(labels_db[valid]).size > 1 else np.nan
ari_db = adjusted_rand_score(y_moons, labels_db)

print("Silhouette (DBSCAN):", round(sil_db, 4))
print("ARI (DBSCAN):", round(ari_db, 4))
print("Noise %:", round((labels_db == -1).mean() * 100, 2))
```

```python
# === Comparison ===
noise_pct = (labels_db == -1).mean() * 100

results = {
    "Best_K": best_K,
    "ARI_KMeans": round(ari_km, 4),
    "ARI_DBSCAN": round(ari_db, 4),
    "Noise_%": round(noise_pct, 2)
}

print("=== Summary ===")
for k, v in results.items():
    print(f"{k}: {v}")
```
