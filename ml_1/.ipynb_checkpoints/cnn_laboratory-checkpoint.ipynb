```python
import os, numpy as np, torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
torch.manual_seed(42)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
use_fallback = False
try:
    from torchvision import datasets, transforms
    tfm = transforms.Compose([transforms.ToTensor()])
    train_ds = datasets.MNIST("./data", train=True,  download=True, transform=tfm)
    test_ds  = datasets.MNIST("./data", train=False, download=True, transform=tfm)
    H, W, n_classes = 28, 28, 10
    train_loader = DataLoader(train_ds, batch_size=128, shuffle=True)
    test_loader  = DataLoader(test_ds,  batch_size=256, shuffle=False)
except Exception:
    use_fallback = True
    from sklearn.datasets import load_digits
    from sklearn.model_selection import train_test_split
    digits = load_digits()
    X = digits.images.astype(np.float32)/16.0  # [n,8,8]
    y = digits.target.astype(np.int64)
    X = X[:, None, :, :]                       # add channel dim: [n,1,8,8]
    from sklearn.model_selection import train_test_split
    X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)
    train_loader = DataLoader(TensorDataset(torch.tensor(X_tr), torch.tensor(y_tr)), batch_size=128, shuffle=True)
    test_loader  = DataLoader(TensorDataset(torch.tensor(X_te), torch.tensor(y_te)), batch_size=256, shuffle=False)
    H, W, n_classes = 8, 8, len(np.unique(y))

print("Fallback:", use_fallback, "| size:", (H,W), "| classes:", n_classes)
```

```python
class SmallCNN(nn.Module):
    def __init__(self, in_ch=1, n_classes=10, small=False):
        super().__init__()
        c1, c2 = (8,16) if small else (16,32)
        self.conv1 = nn.Conv2d(in_ch, c1, 3, padding=1)
        self.conv2 = nn.Conv2d(c1, c2, 3, padding=1)
        self.pool  = nn.MaxPool2d(2,2)
        self.drop  = nn.Dropout(0.25)
        # вычислим выходное число признаков
        dummy = torch.zeros(1,1,H,W)
        with torch.no_grad():
            z = self.pool(F.relu(self.conv1(dummy)))
            z = self.pool(F.relu(self.conv2(z)))
            flat = z.view(1,-1).shape[1]
        self.fc = nn.Linear(flat, n_classes)
    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.drop(x)
        x = x.view(x.size(0), -1)
        return self.fc(x)

model = SmallCNN(in_ch=1, n_classes=n_classes, small=use_fallback).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-4)
```

```python
import matplotlib.pyplot as plt
train_losses, train_accs = [], []
epochs = 8
for ep in range(1, epochs+1):
    model.train(); tot=0; corr=0; run=0.0
    for x, y in train_loader:
        x, y = x.to(device), y.to(device)
        out = model(x)
        loss = criterion(out, y)
        optimizer.zero_grad(); loss.backward(); optimizer.step()
        run += loss.item()*x.size(0)
        corr += (out.argmax(1)==y).sum().item()
        tot  += x.size(0)
    train_losses.append(run/tot); train_accs.append(corr/tot)
    print(f"Epoch {ep:02d}: loss={train_losses[-1]:.4f}, acc={train_accs[-1]*100:.2f}%")
```

```python
model.eval(); tot=0; corr=0
from sklearn.metrics import confusion_matrix, classification_report
all_y, all_p = [], []
with torch.no_grad():
    for x, y in test_loader:
        x, y = x.to(device), y.to(device)
        p = model(x).argmax(1)
        corr += (p==y).sum().item(); tot += x.size(0)
        all_y.extend(y.cpu().numpy()); all_p.extend(p.cpu().numpy())

print("Test accuracy:", round(corr/tot, 4))
print("Confusion matrix:\n", confusion_matrix(all_y, all_p))
print(classification_report(all_y, all_p))

plt.figure(); plt.plot(train_losses); plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("CNN Loss"); plt.show()
plt.figure(); plt.plot(train_accs);  plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.title("CNN Accuracy"); plt.show()
```
