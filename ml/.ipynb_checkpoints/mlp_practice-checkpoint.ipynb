```python
import os, sys, math, random, numpy as np
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset, random_split
torch.manual_seed(42); np.random.seed(42); random.seed(42)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
use_fallback = False
X_train_t = y_train_t = X_test_t = y_test_t = None
try:
    from torchvision import datasets, transforms
    transform = transforms.Compose([transforms.ToTensor()])
    train_ds = datasets.MNIST(root="./data", train=True, download=True, transform=transform)
    test_ds  = datasets.MNIST(root="./data", train=False, download=True, transform=transform)
    train_loader = DataLoader(train_ds, batch_size=128, shuffle=True)
    test_loader  = DataLoader(test_ds, batch_size=256, shuffle=False)
    in_dim = 28*28; n_classes = 10
except Exception as e:
    use_fallback = True
    from sklearn.datasets import load_digits
    from sklearn.model_selection import train_test_split
    digits = load_digits()
    X = digits.images.astype(np.float32) / 16.0  # [n, 8, 8], 0..1
    y = digits.target.astype(np.int64)
    X = X.reshape(len(X), -1)                    # [n, 64]
    X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)
    X_train_t = torch.tensor(X_tr)
    y_train_t = torch.tensor(y_tr)
    X_test_t  = torch.tensor(X_te)
    y_test_t  = torch.tensor(y_te)
    in_dim = X_train_t.shape[1]                  # 64
    n_classes = len(np.unique(y))
    train_loader = DataLoader(TensorDataset(X_train_t, y_train_t), batch_size=128, shuffle=True)
    test_loader  = DataLoader(TensorDataset(X_test_t,  y_test_t),  batch_size=256, shuffle=False)

print("Fallback:", use_fallback, "| input dim:", in_dim, "| classes:", n_classes)
```

```python
class MLP(nn.Module):
    def __init__(self, in_dim, hidden=(256,128), n_classes=10, p_drop=0.0):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, hidden[0]), nn.ReLU(),
            nn.Dropout(p_drop),
            nn.Linear(hidden[0], hidden[1]), nn.ReLU(),
            nn.Dropout(p_drop),
            nn.Linear(hidden[1], n_classes)
        )
    def forward(self, x):
        if x.dim() > 2: x = x.view(x.size(0), -1)
        return self.net(x)

model = MLP(in_dim, hidden=(256,128), n_classes=n_classes, p_drop=0.2).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-4)  # weight_decay = L2
```

```python
import matplotlib.pyplot as plt
train_losses, train_accs = [], []
epochs = 8
for epoch in range(1, epochs+1):
    model.train(); correct=total=0; run_loss=0.0
    for batch in train_loader:
        x, y = batch
        x, y = x.to(device), y.to(device)
        logits = model(x)
        loss = criterion(logits, y)
        optimizer.zero_grad(); loss.backward(); optimizer.step()
        run_loss += loss.item()*x.size(0)
        pred = logits.argmax(1)
        correct += (pred==y).sum().item(); total += x.size(0)
    train_losses.append(run_loss/total); train_accs.append(correct/total)
    print(f"Epoch {epoch:02d}: loss={train_losses[-1]:.4f}, acc={train_accs[-1]*100:.2f}%")
```

```python
model.eval(); correct=total=0
from sklearn.metrics import confusion_matrix, classification_report
all_y, all_p = [], []
with torch.no_grad():
    for x, y in test_loader:
        x, y = x.to(device), y.to(device)
        p = model(x).argmax(1)
        correct += (p==y).sum().item(); total += x.size(0)
        all_y.extend(y.cpu().numpy()); all_p.extend(p.cpu().numpy())
test_acc = correct/total
print("Test accuracy:", round(test_acc, 4))

cm = confusion_matrix(all_y, all_p)
print("Confusion matrix:\n", cm)
print(classification_report(all_y, all_p))
```

```python
plt.figure(); plt.plot(train_losses); plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("MLP Loss"); plt.show()
plt.figure(); plt.plot(train_accs);  plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.title("MLP Accuracy"); plt.show()
```
